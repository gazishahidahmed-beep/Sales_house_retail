### Question 4: Monthly Revenue Calculation (2 Marks)

Calculate the total revenue ('Total Amount') for each month. The result should be a new DataFrame or Series indexed by the month. Display the result.

Calculate and display the cumulative sum of revenue over time. First, group the data by 'Date' and sum the 'Total Amount'. Then, calculate the cumulative sum on this result. Display the last 5 days of the cumulative revenue series.

**Expected Output:** A Series showing the last 5 dates and the cumulative total revenue up to each of those dates.

```python
# Write your answer here
import pandas as pd

# Load the house price dataset
df = pd.read_csv('house_price_india.csv')

# Convert 'Date' from Excel serial number to proper datetime format
df['Date'] = pd.to_datetime(df['Date'], unit='D', origin='1899-12-30')

# Group by 'Date', sum the 'Price', and calculate cumulative sum
cumulative_revenue = df.groupby('Date')['Price'].sum().cumsum()

# Rename the Series to match expected terminology
cumulative_revenue.name = 'Cumulative Total Amount'

# Display the last 5 days
print(cumulative_revenue.tail())
```

### Question 10: Pivot Table for Category vs. Month (2 Marks)

Create a pivot table to show the total 'Total Amount' for each 'Product Category' aggregated by month. First, extract the month number into a 'Month' column. Then create the pivot table with 'Month' as the index and 'Product Category' as columns. Display the resulting pivot table.

**Expected Output:** A table with months (1-12) as rows, product categories as columns, and the total sales for each intersection.

```python
# Write your answer here
import pandas as pd

# Load the house price dataset
df = pd.read_csv('house_price_india.csv')

# Convert the Date column from Excel serial numbers
df['Date'] = pd.to_datetime(df['Date'], unit='D', origin='1899-12-30')

# Extract the month number into a 'Month' column
df['Month'] = df['Date'].dt.month

# Create the pivot table
pivot_table = df.pivot_table(
    values='Price',
    index='Month',
    columns='number of bedrooms',
    aggfunc='sum',
    fill_value=0
)

# Display the resulting pivot table
print(pivot_table)
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
# Write your answer here
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits

# Load data (from Question 11)
digits = load_digits()
X = digits.data

# Initialize the K-Means model with 10 clusters
# random_state is used to ensure reproducible results
kmeans = KMeans(n_clusters=10, random_state=42)

# Fit the model on the features
kmeans.fit(X)

# Assign the resulting cluster labels to the variable
kmeans_labels = kmeans.labels_

# Display the fitted model and a preview of the cluster labels array
print("Fitted Model:\n", kmeans)
print("\nCluster Labels Array (First 20 labels):\n", kmeans_labels[:20])
```

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Write your answer here
import numpy as np
from scipy.stats import mode
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from sklearn.metrics import f1_score

# 1. Load the dataset and fit K-Means (from Questions 11 & 12)
digits = load_digits()
X, y = digits.data, digits.target

kmeans = KMeans(n_clusters=10, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# 2. Map each cluster label to the most frequent true label in that cluster
mapped_labels = np.zeros_like(kmeans_labels)
for i in range(10):
    # Mask to isolate the current cluster 'i'
    mask = (kmeans_labels == i)
    # Find the most common true target label within this cluster
    true_label_mode = mode(y[mask], keepdims=True).mode[0]
    # Assign that true label to the mapped array
    mapped_labels[mask] = true_label_mode

# 3. Calculate the macro-averaged F1 score
macro_f1 = f1_score(y, mapped_labels, average='macro')

# Display the final evaluation score
print(f"Calculated Macro F1 Score: {macro_f1:.4f}")
```
